# Processing conference call data

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
from nltk.tokenize import sent_tokenize
import nltk
#nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.snowball import SnowballStemmer
from nltk.stem import PorterStemmer
stemmer = SnowballStemmer("english")
stop = stopwords.words('english')
ps = PorterStemmer()
from nltk.tokenize import word_tokenize
```

```python
#importing file
path = '/Users/jackphillips/Downloads/'
filename = 'output_csv_combine_2019.csv'

df = pd.read_csv(path+filename, header= None)

```

```python
#taking a look at the data
df.head()
```

```python
#renaming columns
df.columns=['conference','date','name','company','title','group','text']
```

```python
df.head()
```

```python
#learning how to break paragraohs into sentences
text=df['text'][1]
token_text=sent_tokenize(text)
print(token_text)
```

```python
#sentences tokenizer
df['tokenized']=[sent_tokenize(para) for para in df['text']]
    
```

```python
df.head()
```

```python
#working for turning sentences into words
df['tokenized']=[word_tokenize(para) for para in df['sentenences']]
```

```python
#dropping unneeded column
df.drop(['text'],axis=1)
```

```python
# Create lists to fill with values
l_col1 = []
l_col2 = []
l_col3 = []

# iterate over each row and fill our lists
for ix, row in df.iterrows():
    for value in row['tokenized']:
        l_col1.append(value)
        l_col2.append(row['conference'])
        l_col3.append(row['group'])

# Create new dataframe from the two lists
df= pd.DataFrame({'sentenences': l_col1 ,
                         'conference': l_col2 , 'group': l_col3})
#df=df.rename(columns={"sentences":"sents"});
```

```python
#saving new processed df
df.to_csv('/Users/jackphillips/Documents/bysentence.csv')
```
