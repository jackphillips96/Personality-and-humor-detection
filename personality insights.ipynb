Loops to go over each row in each df and run model on. It will append values to the dataframe. 

```python
for index, review in data1['Text'].iteritems():
    
    
        #Enough words to feed into PI
        #Pass on the review to PI
        #service.profile(df1, content_type='text/plain').get_result()
        json_output = service.profile(review,content_type='text/plain').get_result()

        #Iterate over the result (having heirarchy)
        for i in json_output['personality']:
            data1.set_value(index, i['name'], i['percentile']) 
            for j in i['children']:
                #Append the attributes to the data
                data1.set_value(index, j['name'], j['percentile']) 
                        
        for i in json_output['needs']:
            #Append the attributes to the data
            data1.set_value(index, i['name'], i['percentile']) 
            
        for i in json_output['values']:
            #Append the attributes to the data
            data1.set_value(index, i['name'], i['percentile']) 
```
